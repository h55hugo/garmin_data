```python
import sqlite3
import os
print(os.getcwd())
```

```python
db_path = "../db/garmin.db"

# get the new number of activities
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
cursor.execute("SELECT COUNT(*) FROM activities")
n_activity = cursor.fetchone()[0]
conn.close()

print(f"Number of activities in the db: {n_activity}")
```

```python
import pandas as pd

# get the new number of distinct activities
conn = sqlite3.connect(db_path)
df_dist = pd.read_sql_query(
    "SELECT COUNT(DISTINCT activityId) as n_distinct FROM activities", conn)
conn.close()
df_dist
```
